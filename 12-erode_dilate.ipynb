```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

def cv_show(img):
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    plt.show()
```

```python
dige = cv2.imread('01_Picture/05_Dige.png')
cv_show(dige)
```

```python
kernel = np.ones((5,5), np.uint8)
erosion = cv2.erode(dige, kernel, iterations=1)
cv_show(erosion)
```

```python
pie = cv2.imread('01_Picture/06_pie.png')
cv_show(pie)
```

```python
kernel = np.ones((30,30), np.uint8)
erode_1 = cv2.erode(pie, kernel, iterations=1)
erode_2 = cv2.erode(pie, kernel, iterations=2)
erode_3 = cv2.erode(pie, kernel, iterations=3)
res = np.hstack((erode_1, erode_2, erode_3))
cv_show(res)
```

```python
kernel = np.ones((3,3), np.uint8)
dige_erode = cv2.erode(dige, kernel, iterations=1)
cv_show(dige_erode)

kernel = np.ones((3,3), np.uint8)
dige_dilate = cv2.dilate(dige_erode,kernel,iterations=1)
cv_show(dige_dilate)
```

```python
kernel = np.ones((30,30), np.uint8)
dilate_1 = cv2.dilate(pie, kernel, iterations=1)
dilate_2 = cv2.dilate(pie, kernel, iterations=2)
dilate_3 = cv2.dilate(pie, kernel, iterations=3)
res = np.hstack((dilate_1, dilate_2, dilate_3))
cv_show(res)
```
