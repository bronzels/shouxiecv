```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

def cv_show(img):
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    plt.show()
```

```python
vc = cv2.VideoCapture('02_Video/00_Scenery.mp4')
if vc.isOpened():
    open, frame = vc.read()
    print(open)
    cv_show(frame)
else:
    open = False
```

```python
while open:
    ret, frame = vc.read()
    if frame is None:
        break
    if ret == True:
        gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
        cv_show(gray)
    if cv2.waitKey(10) & 0xFF == 27:
        break
vc.release()
```
