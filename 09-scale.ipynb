```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

def cv_show(img):
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    plt.show()
```

```python
img = cv2.imread('01_Picture/01_cat.jpg')
res = cv2.resize(img, (0,0), fx=3, fy=1)
cv_show(res)
```

```python
res = cv2.resize(img, (0,0), fx=1.5, fy=1.5)
cv_show(res)
```
