```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

def cv_show(img):
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    plt.show()
    
def cv2_show(img,name):
    cv2.imshow(name,img)
    cv2.waitKey()
    cv2.destroyAllWindows()
```

```python
img = cv2.imread('01_Picture/08_Car.png')
cv_show(img)
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv_show(gray)
ret, thresh = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY)
cv_show(thresh)
```

```python
contours, hierarchy = cv2.findContours(thresh, cv2.RETR_TREE, cv2.CHAIN_APPROX_NONE)
print(len(contours))
for contour in contours:
    print(contour.shape)
#print(np.array(contours))
print(hierarchy.shape)
```

```python
draw_img = img.copy()
res = cv2.drawContours(draw_img,contours,-1,(0,0,255),2)
cv_show(res)
```

```python
draw_img = img.copy()
res = cv2.drawContours(draw_img,contours,70,(0,0,255),2)
cv_show(res)
```

```python
img = cv2.imread('01_Picture/10_contours.png')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
ret, thresh = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY)

contours, hierarchy = cv2.findContours(thresh, cv2.RETR_TREE, cv2.CHAIN_APPROX_NONE)

cnt = contours[0]
print(cnt)
print(cv2.contourArea(cnt))
print(cv2.arcLength(cnt,True))
```

```python
img = cv2.imread('01_Picture/11_contours2.png')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
ret, thresh = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY)

contours, hierarchy = cv2.findContours(thresh, cv2.RETR_TREE, cv2.CHAIN_APPROX_NONE)
draw_img = img.copy()
res = cv2.drawContours(draw_img, contours, -1, (0,0,255), 2)
cv_show(res)
```

```python
cnt = contours[0]
draw_img = img.copy()
res = cv2.drawContours(draw_img, [cnt], -1, (0,0,255), 2)
cv_show(res)
```

```python
epsilon = 0.1 * cv2.arcLength(cnt, True)
approx = cv2.approxPolyDP(cnt,epsilon,True)
draw_img = img.copy()
res = cv2.drawContours(draw_img, [approx], -1, (0,0,255), 2)
cv_show(res)
```

```python
x,y,w,h = cv2.boundingRect(cnt)
draw_img = img.copy()
draw_img = cv2.rectangle(draw_img, (x,y), (x+w,y+h), (0,244), 2)
cv_show(draw_img)
```

```python
area = cv2.contourArea(cnt)
rect_area = w * h
extent = float(area) /rect_area
print('轮廓面具与边界矩形比：',extent)
```

```python
draw_img = img.copy()
(x,y), redis = cv2.minEnclosingCircle(cnt)
center = (int(x), int(y))
redius = int(redis)
res = cv2.circle(draw_img,center,redius,(0,255,0),2)
cv_show(res)
```
