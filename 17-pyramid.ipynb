```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

def cv_show(img):
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    plt.show()
    
def cv2_show(img,name):
    cv2.imshow(name,img)
    cv2.waitKey()
    cv2.destroyAllWindows()
```

```python
img = cv2.imread('01_Picture/09_AM.png')
cv_show(img)
print(img.shape)
```

```python
up = cv2.pyrUp(img)
cv_show(up)
print(up.shape)
```

```python
down = cv2.pyrDown(img)
cv_show(down)
print(down.shape)
```

```python
up = cv2.pyrUp(up)
cv_show(up)
print(up.shape)
```

```python
up = cv2.pyrUp(img)
up_down = cv2.pyrDown(up)
print(img.shape)
print(img.shape)
cv_show(np.hstack((img, up_down)))
```

```python
down = cv2.pyrDown(img)
down_up = cv2.pyrUp(down)
L_1 = img - down_up
cv_show(L_1)
print(L_1.shape)
```
